```python
import os as os
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# for per-packet analysis
path = './packet_analysis/'
i = 0

for fname in os.listdir(path):
    i += 1
    
    if (os.path.isdir(path + fname)):
        continue
    
    data = np.loadtxt(path + fname)
#     with open(path + fname, 'r') as f:
#         for line in f:
#             data.append(int(line))

#     print(len(data))
    x = np.sort(data)
    y = np.arange(len(x))/float(len(x))
    
    plt.figure(i)
    plt.plot(x, y)
    plt.title(fname)
    plt.xscale('log')
    
#     plt.figure(i)
#     plt.plot(np.sort(data), np.linspace(0, 1, len(data), endpoint=False))
```

```python
# for per-flow analysis
path = './flow_analysis/'
i = 0

for f1 in os.listdir(path + 'tcp/'):
    for f2 in os.listdir(path + 'udp/'): 
        if f1 != f2:
            continue

        
        tcp_data = np.loadtxt(path + 'tcp/' + f1)
        udp_data = np.loadtxt(path + 'udp/' + f2)
        all_data = np.append(tcp_data, udp_data)
#         x = np.sort(data)
#         y = np.arange(len(x))/float(len(x))
        
        i += 1
        plt.figure(i)
        plt.plot(np.sort(tcp_data), np.linspace(0, 1, len(tcp_data), endpoint=False))
        plt.title('tcp_' + f1)
        plt.xscale('log')
        
        i += 1
        plt.figure(i)
        plt.plot(np.sort(udp_data), np.linspace(0, 1, len(udp_data), endpoint=False))
        plt.title('udp_' + f1)
        plt.xscale('log')
        
        i += 1
        plt.figure(i)
        plt.plot(np.sort(all_data), np.linspace(0, 1, len(all_data), endpoint=False))
        plt.title('all_' + f1)
        plt.xscale('log')
```
